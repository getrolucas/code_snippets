```python
import numpy as np
import pandas as pd
```

```python
def readable_number(x):
    """Torna um número de formato padrão em um formato mais legível.
    """
    if x >= 1e6:
        return f"{x*1e-6:1.1f}M"
    return f"{x*1e-3:1.0f}K"


np.random.seed(20)

# criando dataframe para teste
df = pd.DataFrame(
    {'x': np.random.randint(1, 1.1e6, 10)}
).sort_values('x')

# criando nova coluna no dataframe
df['readable_x'] = df.x.apply(readable_number)
df
```
