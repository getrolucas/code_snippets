```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from math import ceil
```

```python
# configuração de estilo
sns.set_style(
    style='darkgrid',                                # grade cinza escura
    rc={'axes.facecolor': '.9', 'grid.color': '.8'}  # ajuste das cores
)

sns.set_palette(palette='deep')                      # pallet de cores dos plots

plt.rcParams["figure.figsize"] = [12, 8]             # tamanho padrão das figs
plt.rcParams["figure.dpi"] = 100                     # resolução das figs
plt.rcParams["figure.facecolor"] = "white"           # plano de fundo
```

```python
# formatando float -> porcentagem com f-strings
x = 1.00002
f'{x:.2%}'
```

```python
# função para multiplos plots
def mutiple_plot(
    df:pd.DataFrame,
    cols:list,
    grid:tuple=None,
    figsize:tuple=(8,6)
) -> None:

    if grid is None:
        size = ceil(df.shape[1]/2)
        grid = (size, size)

    plt.figure(figsize=figsize)
    for i, col in enumerate(cols):
        plt.subplot(grid[0],grid[1],i+1)
        plt.boxplot(df[col], patch_artist=True, whis=1.5) # <- alterar o plot conforme necessário
        plt.title(col, fontdict={'size':18})

    plt.tight_layout(w_pad=6)
    plt.show()
```

```python
# exemplo
df = pd.DataFrame(data={
    'x' : np.random.normal(0, 1, 100),
    'y' : np.random.normal(1, 2, 100) ** 2,
    'z' : np.random.normal(2, 3, 100) ** 3}
)

cols = df.columns

mutiple_plot(df, cols=cols)
```
